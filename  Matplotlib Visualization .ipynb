```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
x = np.linspace(0, 10, 25)
y = x * x+2
fig = plt.figure()                     ## control (left,right,width,height) of the canvas 
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
n = np.array([0,1,2,3,4,5])
fig, axes = plt.subplots(1,4,figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x, x+0.25*np.random.rand(len(x)))
axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)
axes[2].set_title("bar")
axes[2].bar(n,n+2,align="center",alpha=0.9)
axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2,x**3,color='green',alpha=0.5)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0.0,0.0,1,1], polar =True)   ## Radar 
t = np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='blue',lw=3)
```

```python
delta = 0.025                     ## Draw  Contour Image
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize = 10)
ax.set_title('contour map')
```
